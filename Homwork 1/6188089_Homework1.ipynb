# * import PANDAS an read .CSV file

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv ('hw1_std.csv')
print (df)
```

# ☯ START TO CLEAN DATA ☯

# 1. Check null value in each column

```python
df.isnull().sum()
```

# 2. ' sex ' column convert '0 to F' and '1 to M'

```python
df['sex'] = df['sex'].replace({'0': 'F'}) 
df['sex'] = df['sex'].replace({'1': 'M'}) 

print(df)
```

# 3. Replace all missing values ' Nan ' with ' 0 '

```python
df.fillna(value=0, inplace=True)

print(df)
```

# 4. Re-check null value in each column

```python
df.isnull().sum()
```

# 5. Show duplicate rows and eliminate duplicate row

```python
df[df.duplicated(keep=False)==True]
df.drop_duplicates(keep='first')
```

### 2) Find the correlation between all features. Report any strongly related feature pair.
#### 2.1) Use heatmap to  find correlation between all feature

```python
fi, axis = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(727,32,as_cmap=True), square=True, ax=axis)
```

### 3) For two pair of features with the highest correlation score, please create their joint plots. Explain what you observe from the plots.
####  3.1)   1st highest correlation score is {target , ord1} = 0.433798 shown in PINK graph below
#### --->  From a graph can observe that if ord1 = 0 we can assume that they will got NEGATIVE test result for a disease, and if ord1 = 2 we can assume that they will got POSITIVE test result for a disease. But overall, if values of 'ord1' is between 0 to 3 they still can get either POSITIVE or NEGATIVE test result for a disease.

```python
g = sns.jointplot(x='ord1', y='target',data = df, kind='kde', color='m')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels('$ord1$', '$target$')
ax.set_title('target & ord1')
```

####  3.2)   2nd highest correlation score is {target , num3} = 0.373755 shown in ORANGE graph below
#### --->  From a graph can observe that if num3 is around 130 to 154 we can assume that they will got NEGATIVE test result for a disease,  and if num3 is around 155 to 180 we can assume that they will got POSITIVE test result for a disease. But overall, if values of 'num3' is between 71 to 202 they still can get either POSITIVE or NEGATIVE test result for a disease.

```python
g = sns.jointplot(x='num3', y='target',data = df, kind='kde', color='orange')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels('$num3$', '$target$')
ax.set_title('target & num3')
```

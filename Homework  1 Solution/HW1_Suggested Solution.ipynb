## Correlation Heat Map

Next we will plot the correlation value between features as heatmap. (Dataframe has .corr() function for the correlation.)

```python
import seaborn as sns
import numpy as np

fi, axis = plt.subplots(figsize=(10, 8))
corr = in_data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(200,10,as_cmap=True), 
           square=True, ax=axis)
```

```python
corr
```

From the array, the highest correlation value in magnitude is -0.577 which is the correlation between  num4 and ord5. The second highest of such value is -0.436757 which is the correlation between the target and ord4. The highest possitive correlation value is 0.43379826 which is the correlation between the target and ord1. The other values either has no correlation between them, or have low correlation. However, in regard of the "target" column, ord1, ord4, and num4 are the highest correlated variables to it. "sex" and "age" are two other variables with actual names. From the data, they show no correlation between age and sex.

## JointPlot

```python
g = sns.jointplot(x="num4", y="ord5", data = in_data,kind="kde", color="m")
g.plot_joint(plt.scatter, c="r", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$num4$", "$ord5$")
```

The data show that most of the values of ord5 are in category 1 and 2. For ord5 == 2, the num4
usually is in [0,2] range with most of the num4 value concentrated between 0 and 1. For ord5 == 1, the data spread equally from 0 to 4, with the apparent center around num4 value of 2. For ord5 == 0, 
while there are not clear distribution of the data, four apparent outliers for num4 are in this group (as shown below).

```python
plt.figure(figsize=(12,5))
plt.title("Box plot for num4 to check the outliers")
sns.boxplot(x="num4", data = in_data , orient="h", palette = 'pink')
```

```python
g = sns.jointplot(x="target", y="ord4", data = in_data,kind="kde", color="m")
g.plot_joint(plt.scatter, c="r", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$target$", "$ord4$")
```

Most of the target has the disease (target value == 1), the ord4 value is 0. If 
the target does not have the disease, the ord4 are more likely to be 1 than 0. However, 
from the plot, the number of data points in people with no disease, ord4 values of 0 and 
1 are comparable. We may need to use other feature to help classify this data.

Also, if the ord4 value is 1, the patient is likely to be disease-free. And if the ord4 
has the value of 0, then the patient is more likely to have the disease. 
Note however that most ord1 value is 0.

```python
g = sns.jointplot(x="target", y="ord1", data = in_data,kind="kde", color="m")
g.plot_joint(plt.scatter, c="r", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$target$", "$ord1$")
```

For this pair, if the ord1 value is 0, the patient is probably disease free. For everything else, it is most likely that the patient has the disease. Again, most ord1 value is 0.

## Distribution Map


```python
fi= plt.figure(figsize=(14,5))

axis=fi.add_subplot(121)
sns.distplot(in_data[(in_data.target == 0)]['age'],color='b',ax=axis)
axis.set_title('The distribution of the patient without the disease')

axis=fi.add_subplot(122)
sns.distplot(in_data[(in_data.target == 1)]['age'],color='c',ax=axis)
axis.set_title('The distribution of the patient with the disease')
```

As indicated by the correlation value between the target and age, and confirmed by the 
above visualization. The age distribution does not show any definite separation between 
the patients with the dissease and those without. 

```python
fi= plt.figure(figsize=(14,5))

axis=fi.add_subplot(121)
sns.distplot(in_data[(in_data.target == 0)]['sex'],color='b',ax=axis)
axis.set_title('The distribution of the patient without the disease')

axis=fi.add_subplot(122)
sns.distplot(in_data[(in_data.target == 1)]['sex'],color='c',ax=axis)
axis.set_title('The distribution of the patient with the disease')
```

The discrepancy in the group of patient with and without the disease cannot be
observe using the sex variable alone.

```python
sns.catplot(x='age', y='sex', hue='target', kind='bar', palette="inferno", data=in_data)
plt.title('The chance of the patient to have the disease in different sex and age range')
```

Both sex seem to have the same portions of patients with and without the disease. However,
it seems that the group of patient with the disease appear to be younger in this study.

Let us explore it using the following plots. Then we will discuss it at the end.

```python
sns.catplot(x="sex", y="age", hue="target", kind="violin", palette = 'inferno', data=in_data)
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(y="sex", x="age", data = in_data[(in_data.target == 1)] 
            ,orient="h", palette = 'inferno')
plt.title("The box plot of patient with disease")
```

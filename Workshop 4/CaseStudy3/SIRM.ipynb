## Simple Calculation following the studies of Covid 19 on the plane [3]

For the BTS train, each ride has 4 cars. Each car is 87-meter long and 3-meter wide [4]. Based on the number reported by BTS [3], we calculated that there are about 430 trains per day. BTS services 672000 riders per day on average. For simplicity, let us assume uniform distribution for all senarios in this simulation. There are 48 stations in two lines, this implies that each station services about 14000 departing riders, and another 14000 arriving riders. This implies that there are about 65 riders per train getting on at each station. Let us assume that each person rides the BTS for 6 stations on average, and there is an infected person in that train ride. At peak, there are about 325 riders per train, 82 riders per car at any given time. With 42 seats, there are 40 stading riders. If one person stands in the middle of two opporsite seats, there are 19 riders left in the doors area. This implies about 6 person in each doors area. If the infected person stand in the middle, there can be about 14 riders within the radius of transmission. For 6 stations, there are 14*5 = 70 riders being in transmissed distance to an infected rider. Hence, if you ride on the same train, there are  0.045 probability of being in transmisted distance of the infected person. Assuming uniform distribution, if you ride once per day, there are 0.0001 probability of being in the transmited distance of the infected person. Note that, the uniform distribution assumption is unrealistic.

# Covid-19 simulation with SIR model

For this study, we create a simulation based on a compartmental model. In a compartmental model, the population is divided into different groups. The model that we ues in this simulation is called the SIR model. The SIR model separates the population into three groups: the susceptible (S), the infected (I), and the removed (R). The code for this study is taken from [6].

For this model, the size of each group is represented by a function of time, t. Specifically,

S(t) is the number of individuals who are susceptible to the disease on the train at time t.
I(t) is the number of individuals who are capable of spreading the disease to those in the susceptible group.
R(t): the number of individuals who have been infected and then removed from the rest of the group.

Each function is changed based on the transition rates. The transition rate must be specified for all pairwises group. For example, the rate of population growth from S(t) to S(t+x). The rate of disease spreading which implies the transition from a fraction of S(t) to I(t+x).

Note that for our current model, we are simply consider Si,t to be n-1, while Ii,t = 1, Ri,t = 0

For a particular close space, one can use the differential equation of the SIR model. Let us assume that, in the beginning, there are 15000 people, with 2 infectious person in the group. 
For Covid 19, the beta and the gamma are estimated to based on the data as described in [6]. Note that, the simulation taken from Sasaki was based on the real world data. 

```python
import numpy as np
import pandas as pd
from scipy.integrate import solve_ivp
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from datetime import timedelta, datetime

S_0 = 15000
I_0 = 2
R_0 = 0

START_DATE = {
  'Japan': '1/22/20',
  'Italy': '1/31/20',
  'Republic of Korea': '1/22/20',
  'Iran (Islamic Republic of)': '2/19/20'
}

class Learner(object):
    def __init__(self, country, loss):
        self.country = country
        self.loss = loss

    def load_confirmed(self, country):
      df = pd.read_csv('time_series_19-covid-Confirmed.csv')
      country_df = df[df['Country/Region'] == country]
      return country_df.iloc[0].loc[START_DATE[country]:]

    def load_recovered(self, country):
      df = pd.read_csv('time_series_19-covid-Recovered.csv')
      country_df = df[df['Country/Region'] == country]
      return country_df.iloc[0].loc[START_DATE[country]:]

    def extend_index(self, index, new_size):
        values = index.values
        current = datetime.strptime(index[-1], '%m/%d/%y')
        while len(values) < new_size:
            current = current + timedelta(days=1)
            values = np.append(values, datetime.strftime(current, '%m/%d/%y'))
        return values

    def predict(self, beta, gamma, data, recovered, country):
        predict_range = 150
        new_index = self.extend_index(data.index, predict_range)
        size = len(new_index)
        def SIR(t, y):
            S = y[0]
            I = y[1]
            R = y[2]
            return [-beta*S*I, beta*S*I-gamma*I, gamma*I]
        extended_actual = np.concatenate((data.values, [None] * (size - len(data.values))))
        extended_recovered = np.concatenate((recovered.values, [None] * (size - len(recovered.values))))
        return new_index, extended_actual, extended_recovered, solve_ivp(SIR, [0, size], [S_0,I_0,R_0], t_eval=np.arange(0, size, 1))

    def train(self):
        data = self.load_confirmed(self.country)
        recovered = self.load_recovered(self.country)
        optimal = minimize(loss, [0.001, 0.001], args=(data, recovered), method='L-BFGS-B', bounds=[(0.00000001, 0.4), (0.00000001, 0.4)])
        print(optimal)
        beta, gamma = optimal.x
        new_index, extended_actual, extended_recovered, prediction = self.predict(beta, gamma, data, recovered, self.country)
        df = pd.DataFrame({'Confirmed': extended_actual, 'Recovered': extended_recovered, 'S': prediction.y[0], 'I': prediction.y[1], 'R': prediction.y[2]}, index=new_index)
        fig, ax = plt.subplots(figsize=(15, 10))
        ax.set_title(self.country)
        df.plot(ax=ax)
        print(f"country={self.country}, beta={beta:.8f}, gamma={gamma:.8f}, r_0:{(beta/gamma):.8f}")

def loss(point, data, recovered):
    size = len(data)
    beta, gamma = point
    def SIR(t, y):
        S = y[0]
        I = y[1]
        R = y[2]
        return [-beta*S*I, beta*S*I-gamma*I, gamma*I]
    solution = solve_ivp(SIR, [0, size], [S_0,I_0,R_0], t_eval=np.arange(0, size, 1), vectorized=True)
    l1 = np.sqrt(np.mean((solution.y[1] - data)**2))
    l2 = np.sqrt(np.mean((solution.y[2] - recovered)**2))
    alpha = 0.1
    return alpha * l1 + (1 - alpha) * l2

learner = Learner('Japan', loss)
learner.train()
learner = Learner('Republic of Korea', loss)
learner.train()
learner = Learner('Italy', loss)
learner.train()
learner = Learner('Iran (Islamic Republic of)', loss)
learner.train()
```
